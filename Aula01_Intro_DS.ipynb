```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

lista_idades = [12, 41, 76, 10, 52, 4, 32, 28, 8, 14]

def maiores_de_18(lista_idades):
  return [idade for idade in lista_idades if idade >= 18]

result = maiores_de_18(idades)

print(result)

def dicionario_maiores(lista_idades):
  maiores = [idade for idade in lista_idades if idade >= 18]

  dic = {i: idade for i, idade in enumerate(maiores)}
  return dic

result2 = dicionario_maiores(idades)
print(result2)
```
