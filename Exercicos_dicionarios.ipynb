```python
idades = [15, 22, 17, 30, 14, 19, 40]

maiores_lista = []
maiores_dict = {}
contador = 0
pessoa = 1

for n in idades:
    if n > 18:
        maiores_lista.append(n)
        maiores_dict[f"pessoa{pessoa}"] = n
        contador += 1
        pessoa += 1

print("Quantidade:", contador)
print("Lista:", maiores_lista)
print("Dicionário:", maiores_dict)
```
