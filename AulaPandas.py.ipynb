```
idades = [16, 25, 19, 14, 32]

maiores = []
contador = 0
for n in idades:
  if n > 18:
    maiores.append(n)
    contador += 1

print(maiores)
print(contador)
```

```
import pandas as pd

dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Duda"],
    "idade": [23, 17, 35, 16]
}


df = pd.DataFrame(dados)


maiores = df[df["idade"] > 18]
print(maiores)
print(len(maiores))
```
